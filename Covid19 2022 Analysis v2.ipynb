```python
import pandas as pd
import plotly.express as px
import requests
import io
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
r = requests.get(url).content

df = pd.read_csv(io.StringIO(r.decode("utf-8")))
df
```

```python
#df.head()
#df.tail()
df.info()
#df.describe()
```

```python
# Convert date from object to datetime64
df['date'] = pd.to_datetime(df.date)
```

```python
# Select countries. Might consider having DropDown Lists in next iteration.
countries = ['New Zealand', 'Malaysia']
df = df[df['location'].isin(countries)]
```

```python
fig = px.line(df, x="date", y="new_cases", color="location", title="C19 New Cases across time")
fig.show()
```

```python
fig = px.line(df, x="date", y="new_cases_per_million", color="location", title="C19 New Cases across time per million people")
fig.show()
```
